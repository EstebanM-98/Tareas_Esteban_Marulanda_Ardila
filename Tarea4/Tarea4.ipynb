```python
import numpy as np
import pandas as pd
```

```python
#1)

exo=pd.read_csv("exoplanets.csv") #Cargaos los datos
exo=pd.DataFrame(exo)  #Creamos un DF


names=['NAME','TEFF','MASS','A','DENSITY','R','STAR','MSTAR','RSTAR','BINARY']  #Nombres de las columnas requeridas

exo1=pd.DataFrame() #Contendra columnas de exo con los nombres de names

for i in range(0,len(names)):
    exo1[names[i]]=exo[names[i]]
    
#Eliminare aquellos planetas que tiene alguna propiedad con 'NaN' ya que para efectos de computo no serviría
    
exo1=exo1.dropna() #Borra los planetas que tengan algún parametro con 'NaN'

exo1.index=[i for i in range(0,len(exo1))] #Renombramos las etiquetas de las filas

exo1
```

```python
#2. Aqui descartare los que tienen BINARY==1,  ya que el valor '0' se entiende como que no es un sistema binario

for i in range(0,len(exo1)):
    
    fila = exo1.loc[i]
    
    if fila[9]==1   :
        exo1=exo1.drop(i) #Elimina los sistemas binarios 

exo1.index=[i for i in range(0,len(exo1))] #Renombramos las etiquetas de las filas
exo1  #Sistemas no binarios
```

```python
#3)
#Para obtener la masa joviana en masas terrestres multiplicamos MSTAR por 317,83

MASSE=[]

for i in range(0,len(exo1)):
    MASSE.append(317.83*exo1['MSTAR'][i])
    
exo1['MASSE']=pd.Series(MASSE)  #Añadimos columna a exo1 con el nombre de MASSE

exo1
```

```python
#4)

#1 radio de jupiter = 11,209 Radio de la tierra 

RE=[]

for i in range(0,len(exo1)):
    RE.append(11.209*exo1['RSTAR'][i])
    
exo1['RE']=pd.Series(RE)  #Añadimos columna a exo1 con el nombre de MASSE

exo1
```

```python
#5)
#Añadir columna con la luminosidad
ste_bolt=5.68e-8  #Constante de Steffan-Boltzman en W/(m^2*K^4)

ste_boltRT=ste_bolt/(1.567e-7)**2 #Constnte de Steffan-Boltzman en W/(R_T^2*K^4) donde R_T son radios terrestres

LUM=[] #Contendra luminosidad en W

for i in range(0,len(exo1)):
    LUM.append(4*np.pi*ste_boltRT*(exo1['RE'][i])**2*(exo1['TEFF'][i])**4) 
    
exo1['LUM']=pd.Series(LUM)  #Añadimos columna a exo1 con el nombre de LUM en

exo1
```

```python
#6)

#Añadir columna ri
a_0=1.378e-4
r_is=0.72
T_s=5780.
b_i=3.809e-9
r_os=1.77
a_i=2.7619e-5
b_0=1.428e-9
factor=3.828e26 #Factor que convierte la luminosidad de 'W' a unidades solares.

ri=[(r_is-a_i*(exo1['TEFF'][i]-T_s)-b_i*(exo1['TEFF'][i]-T_s)**2)*(exo1['LUM'][i]/factor)**0.5  for i in range(0,len(exo1))] #Contendra luminosidad

    
exo1['ri']=pd.Series(ri)  #Añadimos columna a exo1 con el nombre de LUM

exo1
```

```python
#Añadir columna ro


ro=[] #Contendra luminosidad

ro=[(r_os-a_0*(exo1['TEFF'][i]-T_s)-b_0*(exo1['TEFF'][i]-T_s)**2)*(exo1['LUM'][i]/factor)**0.5 for i in range(0,len(exo1))] #Contendra luminosidad
    
exo1['ro']=pd.Series(ro)  #Añadimos columna a exo1 con el nombre de LUM

exo1
```

```python
#7)

# Filtro para eliminar planetas cuya densidad sea menor a 5

densidad = exo1.query('`DENSITY`>5')
densidad
```

```python
#7) Filtro sin funciones

for i in range(0,len(exo1)):
    
    fila = exo1.loc[i]
    
    if fila[4]< 5.0 :
        exo1=exo1.drop(i) #Elimina los planetas  que tengan una densidad inferior a las 5
    
exo1.index=[i for i in range(0,len(exo1))] #Renombramos las etiquetas de las filas
exo1
```

8) 

De la referencia tenemos la formula para calcular el parametro 'HZD'. El r que se presenta en esta ecuación se asumira que es el semieje mayor presentado en la tabla ('A'), debemos por tanto ver que planetas cumplen la condición de que -1<HZD<1 .

Our approach to convert the HZ into an analog scale, and not just a yes-no condition, consist in calculating the distance of the exoplanet from the middle point of the HZ and normalized to half-width of the HZ. This in effect creates a new metric or distance function in habitable space and habitable zone units (HZU). The habitable zones distance (HZD) is therefore given as

$$HZD=\frac{2r-r_{0}-r_{i}}{r_{0}-r_{i}}$$

where r is the distance of the exoplanet from the star in AU units and HZd is the HZD in HZU units. These units are very practical because they mean the same thing independently of the stellar system under consideration. HZD values between -1 and +1 HZU always correspond to planets within the HZ.

```python
#Añadir columna con HZD


HZD=[] #Contendra HZD

HZD=[ (2*exo1['A'][i]-exo1['ro'][i]-exo1['ri'][i])/(exo1['ro'][i]-exo1['ri'][i]) for i in range(0,len(exo1))]
    
exo1['HZD']=pd.Series(HZD)  #Añadimos columna a exo1 con el nombre de HZD

exo1
```

```python
#Fltro con query

ZONA = exo1.query('-1<`HZD`<1')
ZONA
```
